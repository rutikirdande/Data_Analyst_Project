```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('Electric_Vehicle_Population_Data.csv')
```

```python
# Handling Missing Values

# Fill missing values in 'Electric Range' with the median
median_electric_range = df['Electric Range'].median()
df['Electric Range'].fillna(median_electric_range, inplace=True)

# Fill missing values in 'Model Year' with the median
median_model_year = df['Model Year'].median()
df['Model Year'].fillna(median_model_year, inplace=True)

# Fill missing values in 'Legislative District' with the median
median_legislative_district = df['Legislative District'].median()
df['Legislative District'].fillna(median_legislative_district, inplace=True)

# Handling Duplicate Values

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Data Type Conversion

# Convert 'Model Year' to integer
df['Model Year'] = df['Model Year'].astype(int)

# Remove Irrelevant Columns

# Drop columns that are not needed for analysis
columns_to_drop = ['VIN (1-10)', 'DOL Vehicle ID', '2020 Census Tract', 'Vehicle Location']
df.drop(columns=columns_to_drop, inplace=True)

# Handling Categorical Variables

# Convert categorical variables to lowercase (example: 'County')
df['Country'] = df['Country'].str.lower()

# Remove leading/trailing spaces from categorical columns (example: 'City')
df['City'] = df['City'].str.strip()


# Identify and handle outliers in numerical columns using z-scores or other methods

```

```python
# Identify Outliers using Z-Scores
from scipy.stats import zscore
# Specify the numerical columns you want to check for outliers
numerical_columns = ['Electric Range', 'Model Year', 'Legislative District']

# Calculate z-scores for the specified columns
z_scores = np.abs(zscore(df[numerical_columns]))

# Define a z-score threshold for identifying outliers (e.g., 3 standard deviations)
z_threshold = 3

# Identify rows with outliers
outliers = (z_scores > z_threshold).any(axis=1)

# Handling Outliers

# Replace outliers with the median value of the respective column
for column in numerical_columns:
    median_value = df[column].median()
    df.loc[outliers, column] = median_value
```

```python
# Exploratory Data Analysis

# Overview of the first few rows
print(df.head())

# Summary statistics
print(df.describe())

# Correlation Analysis

# Compute correlation matrix
correlation_matrix = df.corr()
```

```python
# Plot correlation matrix heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.tight_layout()
plt.show()
```

```python
# Grouping and Aggregation

# Group by 'Make' and calculate average 'Electric Range'
avg_range_by_make = df.groupby('Make')['Electric Range'].mean().sort_values(ascending=False)
```

```python
# Plot bar chart for average electric range by make
plt.figure(figsize=(12, 6))
avg_range_by_make.plot(kind='bar')
plt.title('Average Electric Range by Make')
plt.xlabel('Make')
plt.ylabel('Average Electric Range')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Value counts for 'Electric Vehicle Type'
plt.figure(figsize=(8, 6))
df['Electric Vehicle Type'].value_counts().plot(kind='bar')
plt.title('Distribution of Electric Vehicle Types')
plt.xlabel('Electric Vehicle Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Comparisons and Analysis Steps for Specific Columns

# Compare Electric Range for different Electric Vehicle Types
plt.figure(figsize=(10, 6))
sns.boxplot(x='Electric Vehicle Type', y='Electric Range', data=df)
plt.title('Electric Range by Electric Vehicle Type')
plt.xlabel('Electric Vehicle Type')
plt.ylabel('Electric Range')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Compare Electric Range by Model Year
plt.figure(figsize=(12, 6))
sns.boxplot(x='Model Year', y='Electric Range', data=df)
plt.title('Electric Range by Model Year')
plt.xlabel('Model Year')
plt.ylabel('Electric Range')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Analyze Clean Alternative Fuel Vehicle (CAFV) Eligibility
plt.figure(figsize=(8, 6))
df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts().plot(kind='bar')
plt.title('Clean Alternative Fuel Vehicle Eligibility')
plt.xlabel('CAFV Eligibility')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
